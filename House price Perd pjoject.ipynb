```python
#address, date, postcode, YearBuilt, lattitude, longtitude columns will not be that helpful 
#to predict the house price so we can drop them to avoid overfitting
df.drop(['Address','Date','Postcode','YearBuilt','Lattitude','Longtitude'],axis=1,inplace=True)
```

```python
df.isnull().sum()
```

##### there are more than 50% nulls presemt in data so we have to fill them 

```python
#Fill the null value of property count, distance, Bedroom2, Bathroom, Car with 0
df[['Propertycount','Distance','Bedroom2','Bathroom','Car']]=df[['Propertycount','Distance','Bedroom2','Bathroom','Car']].fillna(0)
```

```python
#Fill Null value of land size and bidding area columns with Mean 
df[['Landsize']]=df[['Landsize']].fillna(df.Landsize.mean())
df[['BuildingArea']]=df[['BuildingArea']].fillna(df.BuildingArea.mean())
```

```python
#creating dummy dataframe to avoid categorical value 
df=pd.get_dummies(df,drop_first=True)
```

```python
data=df.drop(['Rooms','Price','Distance','Bedroom2','Bathroom','Car','Landsize','BuildingArea','Propertycount'],axis=1)
```

##### spliting the data into x & y 


```python
x=df.drop('Price',axis=1)
y=df['Price']
```

```python
x.shape
```

```python
y.shape
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.7,random_state=2)
```

## creating linear regression model 

```python
from sklearn.linear_model import LinearRegression
```

```python
model=LinearRegression()
model.fit(x_train,y_train)
```

```python
model.score(x_train,y_train)
```

```python
model.score(x_test,y_test)
```

## Ridge Regression 

```python
from sklearn.linear_model import Ridge
```

```python
reg=Ridge(alpha=10,max_iter=100)
reg.fit(x_train,y_train)
```

```python
reg.score(x_train,y_train)
```

```python
reg.score(x_test,y_test)
```

### Lasso Regression

```python
from sklearn.linear_model import Lasso
```

```python
las=Lasso(alpha=50,max_iter=100)
las.fit(x_train,y_train)
```

```python
las.score(x_train,y_train)
```

```python
las.score(x_test,y_test)
```

#### Hyper-parameter tuning

```python
reg_model.score(x_test,y_test)
```

```python
y_pred=reg_model.predict(x_test)
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error,r2_score
```
